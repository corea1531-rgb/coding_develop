```python
import requests
import json
import calendar
from datetime import datetime , timedelta

today = datetime.today()

year = today.year
month = today.month

from_date = (today - timedelta(days=15)).strftime('%Y-%m-%d')
to_date = (today + timedelta(days=15)).strftime('%Y-%m-%d')


headers = {
    'accept': 'application/json, text/plain, */*',
    'accept-language': 'ko-KR,ko;q=0.9,zh-CN;q=0.8,zh;q=0.7,en-US;q=0.6,en;q=0.5,ja;q=0.4',
    'charset': 'utf-8',
    'origin': 'https://m.sports.naver.com',
    'priority': 'u=1, i',
    'referer': 'https://m.sports.naver.com/kbaseball/schedule/index?category=kbo&date=2026-03-12',
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/146.0.0.0 Safari/537.36',
    'x-sports-backend': 'kotlin',
}

params = {
    'fields': 'basic,schedule,baseball,manualRelayUrl',
    'upperCategoryId': 'kbaseball',
    'categoryId': 'kbo',
    'fromDate': from_date,
    'toDate': to_date,
    'roundCodes': '',
    'size': '500',
}


response = requests.get(
    'https://api-gw.sports.naver.com/schedule/games',
    params=params,
    headers=headers
)

games = response.json()['result']['games']

for game in games:
    game_id = game.get('gameId', '')
    round_code = game.get('roundCode','')
    game_date = game.get('gameDate', '')
    home_team = game.get('homeTeamName', '')
    away_team = game.get('awayTeamName', '')
    home_starting = game.get('homeStarterName','')
    away_starting = game.get('awayStarterName','')
    status_code = game.get('statusCode', '')
    winner = game.get('winner', '')

    # print(game_id, round_code,game_date, home_team, away_team,home_starting,away_starting, home_score, away_score, winner, status_code, sep='\n')
    # print('=' * 20)

    if status_code == 'RESULT':
        record_url = f'https://api-gw.sports.naver.com/schedule/games/{game_id}/record'
        record_response = requests.get(record_url,headers=headers)
        record_data = record_response.json()['result']['recordData']

        away_inning_scores = record_data.get('scoreBoard',{}).get('inn',{}).get('away',[])
        home_inning_scores = record_data.get('scoreBoard',{}).get('inn',{}).get('home',[])
        away_inning_1_scores = away_inning_scores[0]
        home_inning_1_scores = home_inning_scores[0]
        away_score_to_4 = sum(away_inning_scores[:4])
        home_score_to_4 = sum(home_inning_scores[:4])
        away_score_to_5 = sum(away_inning_scores[:5])
        home_score_to_5 = sum(home_inning_scores[:5])
        away_hits = record_data.get('scoreBoard',{}).get('rheb',{}).get('away',{}).get('h','')
        home_hits = record_data.get('scoreBoard',{}).get('rheb',{}).get('home',{}).get('h','')
        away_total_ball = record_data.get('scoreBoard', {}).get('rheb', {}).get('away', {}).get('b', '')
        home_total_ball = record_data.get('scoreBoard', {}).get('rheb', {}).get('home', {}).get('b', '')
        away_starting_pitcher = record_data.get('pitchersBoxscore', {}).get('away', [])[0].get('name', '')
        home_starting_pitcher = record_data.get('pitchersBoxscore', {}).get('home', [])[0].get('name', '')
        
        print(away_total_ball,away_starting_pitcher)
       
        break
```
