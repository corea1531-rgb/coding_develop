```python
# 화이트 색상으로 출력 

from pathlib import Path
from tkinter import Tk, filedialog
from rembg import remove
from PIL import Image , ImageEnhance
import io
import time
import shutil




def select_images():
    root = Tk()
    root.withdraw()

    file_paths = filedialog.askopenfilenames(
        title="누끼 딸 이미지 선택",
        filetypes=[
            ("이미지 파일", "*.png;*.jpg;*.jpeg;*.webp;*.bmp"),
            ("모든 파일", "*.*"),
        ],
    )
    root.destroy()
    return list(file_paths)


def remove_background_and_save_as_jpg(image_path):
    image_path = Path(image_path)

    output_dir = image_path.parent / "output"
    failed_dir = image_path.parent / "failed"

    output_dir.mkdir(exist_ok=True)
    failed_dir.mkdir(exist_ok=True)

    output_path = output_dir / f"{image_path.stem}.jpg"

    start = time.time()

    try:
        # 원본 읽기
        with open(image_path, "rb") as f:
            input_data = f.read()

        # rembg 실행 (결과는 보통 투명 PNG 바이트)
        output_data = remove(input_data)

        # 바이트 데이터를 이미지로 열기
        result_img = Image.open(io.BytesIO(output_data)).convert("RGBA")

        # 흰 배경 만들기
        white_bg = Image.new("RGB", result_img.size, (255, 255, 255))

        # 알파 채널을 마스크로 사용해서 흰 배경 위에 합성
        white_bg.paste(result_img, mask=result_img.split()[3])

        # JPG 저장
        white_bg.save(output_path, "JPEG", quality=100, subsampling=0)
        
        elapsed = time.time() - start
        print(f"완료: {output_path.name} / {elapsed:.2f}초")
        return True

    except Exception as e:
        shutil.copy2(image_path, failed_dir / image_path.name)
        print(f"실패: {image_path.name}")
        print("오류:", e)
        return False


def main():
    image_paths = select_images()

    if not image_paths:
        print("선택된 파일이 없습니다.")
        return

    total = len(image_paths)
    success_count = 0
    fail_count = 0

    print(f"총 {total}개 파일 선택됨")

    for idx, image_path in enumerate(image_paths, start=1):
        print(f"[{idx}/{total}] 처리 중: {image_path}")
        ok = remove_background_and_save_as_jpg(image_path)

        if ok:
            success_count += 1
        else:
            fail_count += 1

    print("-" * 40)
    print("작업 완료")
    print(f"성공: {success_count}개")
    print(f"실패: {fail_count}개")


if __name__ == "__main__":
    main()
```

```python
# 화이트 색상으로 출력 , 화질개선

from pathlib import Path
from tkinter import Tk, filedialog
from rembg import remove
from PIL import Image ,ImageEnhance
import io
import time
import shutil


def select_images():
    root = Tk()
    root.withdraw()

    file_paths = filedialog.askopenfilenames(
        title="누끼 딸 이미지 선택",
        filetypes=[
            ("이미지 파일", "*.png;*.jpg;*.jpeg;*.webp;*.bmp"),
            ("모든 파일", "*.*"),
        ],
    )
    root.destroy()
    return list(file_paths)


def remove_background_and_save_as_jpg(image_path):
    image_path = Path(image_path)

    output_dir = image_path.parent / "output"
    failed_dir = image_path.parent / "failed"

    output_dir.mkdir(exist_ok=True)
    failed_dir.mkdir(exist_ok=True)

    output_path = output_dir / f"{image_path.stem}.jpg"

    start = time.time()

    try:
        # 원본 읽기
        with open(image_path, "rb") as f:
            input_data = f.read()

        # rembg 실행 (결과는 보통 투명 PNG 바이트)
        output_data = remove(input_data)

        # 바이트 데이터를 이미지로 열기
        result_img = Image.open(io.BytesIO(output_data)).convert("RGBA")

        # 흰 배경 만들기
        white_bg = Image.new("RGB", result_img.size, (255, 255, 255))

        # 알파 채널을 마스크로 사용해서 흰 배경 위에 합성
        white_bg.paste(result_img, mask=result_img.split()[3])

        enhancer = ImageEnhance.Sharpness(white_bg)
        white_bg = enhancer.enhance(1.2)

        # JPG 저장
        white_bg.save(output_path, "JPEG", quality=100, subsampling=0)
        
        elapsed = time.time() - start
        print(f"완료: {output_path.name} / {elapsed:.2f}초")
        return True

    except Exception as e:
        shutil.copy2(image_path, failed_dir / image_path.name)
        print(f"실패: {image_path.name}")
        print("오류:", e)
        return False


def main():
    image_paths = select_images()

    if not image_paths:
        print("선택된 파일이 없습니다.")
        return

    total = len(image_paths)
    success_count = 0
    fail_count = 0

    print(f"총 {total}개 파일 선택됨")

    for idx, image_path in enumerate(image_paths, start=1):
        print(f"[{idx}/{total}] 처리 중: {image_path}")
        ok = remove_background_and_save_as_jpg(image_path)

        if ok:
            success_count += 1
        else:
            fail_count += 1

    print("-" * 40)
    print("작업 완료")
    print(f"성공: {success_count}개")
    print(f"실패: {fail_count}개")


if __name__ == "__main__":
    main()
```
