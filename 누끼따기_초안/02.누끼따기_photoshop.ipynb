```python
from tkinter import Tk, filedialog
from pathlib import Path
import win32com.client
import time

root = Tk()
root.withdraw()

file_paths = filedialog.askopenfilenames(
    title="포토샵에서 열 파일들 선택",
    filetypes=[("이미지 파일", "*.png;*.jpg;*.jpeg;*.bmp;*.webp;*.psd")]
)
root.destroy()

if not file_paths:
    print("파일 선택 안 함")
else:
    ps = win32com.client.Dispatch("Photoshop.Application")
    ps.Visible = True

    total = len(file_paths)

    for idx, file_path in enumerate(file_paths, start=1):
        image_path = Path(file_path).resolve()

        print(f"[{idx}/{total}] 처리 중: {image_path.name}")
        print("exists:", image_path.exists())
        print("absolute:", image_path)

        start_time = time.time()

        try:
            doc = ps.Open(str(image_path))
            print("열기 성공:", doc.Name)

            # 포토샵 액션 실행
            ps.DoAction("누끼_jpg", "Python")

            elapsed = time.time() - start_time
            print(f"액션 실행 완료 / 소요시간: {elapsed:.2f}초")

            time.sleep(1)

        except Exception as e:
            elapsed = time.time() - start_time
            print("실패:", image_path.name)
            print("오류:", e)
            print(f"실패까지 소요시간: {elapsed:.2f}초")
```
